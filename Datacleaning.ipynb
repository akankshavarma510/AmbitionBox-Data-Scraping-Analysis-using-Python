```python
#importing required modules
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import seaborn as sns
import time
```

```python
# Reading CSV file into dataframe
dataframe = pd.read_csv("CompanyList.csv")
dataframe
```

```python
"""drops complete row if Founded Year consists any invalid data like in few rows we have 10k-50k
which is invalid, by finding founded year with contains '-' """
dataframe.drop(dataframe[dataframe['Founded Year'].str.contains('-',na=False)].index)
```

```python
#drops the rows if elements from CVS file of founded year are invalid
invalid = dataframe[dataframe['Founded Year'].str.contains('-',na=False)].index
dataframe.drop(invalid,inplace=True)
```

```python
#fills the missing values in founded year column by forward filling method
dataframe['Founded Year'] = dataframe['Founded Year'].fillna(method='pad')

# seperate columns are created with names Years old and founded year
dataframe['Years old']=dataframe['Founded Year'].apply(lambda x:" ".join(re.findall('\s\((.*?)\s[yrs]',str(x))))
#dataframe['Years old']=dataframe['Founded Year'].apply(lambda x:int(x[x.find('(')+1:x.find('yrs')].strip()) if type(x)==type('x') else np.NaN)
dataframe['Founded Year']=dataframe['Founded Year'].apply(lambda x:" ".join(re.findall('(\d{4})',str(x))))
```

```python
#fills the missing values in Highly rated column by forward filling method
dataframe['Highly rated'] = dataframe['Highly rated'].fillna(method='pad')
```

```python
# converts lakhs into 100000 and k into 1000
def Count(con:pd.core.frame.DataFrame):
    for i in con:  
        try:
            c = (str(i).lower().replace(' lakh+','00000').replace(' lakh','00000').replace('k','000').split('-'))
            return int(max(c))
        except Exception as e:
            return np.NaN
dataframe['Global Count']=Count(dataframe['Global Count'])
dataframe['India Count'] = Count(dataframe['India Count'])

# fills missing value with mean

dataframe['Global Count'] = dataframe['Global Count'].fillna(dataframe['Global Count'].mean())
dataframe['India Count'] = dataframe['India Count'].fillna(dataframe['India Count'].mean())
```

```python
## converting string literals to int
dataframe['Rating'].astype(float)
dataframe['Founded Year'].astype(int)
dataframe['Years old'].astype(float)
```
