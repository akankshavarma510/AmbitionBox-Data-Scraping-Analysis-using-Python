```python
## importing required modules
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
## creating lists to store extracted data
Company_Name = []
Industry = []
Rating = []
Highly_rated = []
Critically_rated = []
#Page_Number = []

## loops through first 25 pages
for i in range(1,26):
    # defining url
    url  = f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={i}"
    
    # sending get request to url
    response=requests.get(url,headers={'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Mobile Safari/537.36'})
    
    # Parsing the page content
    page_content = response.text
    soup = BeautifulSoup(page_content,'html.parser')

    #finding all companyCardWrapper class elements of individual page
    for var in soup.find_all('div',attrs={'class':'companyCardWrapper'}):
        
        # Extracting required data
        Company_Name1 = var.find('h2',attrs={'class':'companyCardWrapper__companyName'})
        Industry1 = var.find('span',attrs={'class':'companyCardWrapper__interLinking'})
        #Employees1 = x.find('span',attrs={'class':'companyCardWrapper__interLinking'})
        Rating1 = var.find('div',attrs={'class':'companyCardWrapper__header'})
        Highly_rate1 = var.find('span',attrs={"class":"companyCardWrapper__ratingValues"})
        critical_rating_span = var.find('span', class_='companyCardWrapper__ratingHeader--critical')
        if critical_rating_span:
            rating_value_span = critical_rating_span.find_next('span', class_='companyCardWrapper__ratingValues')
        

        # Appending the extracted data to the respective lists
        if Company_Name1 is None:
            Company_Name.append(np.nan)
        else:
            Company_Name.append(Company_Name1.text.strip())
        if Industry1 is None:
            Industry.append(np.nan)
        else:
            line = Industry1.text.strip()
            Industry.append(line.split('|')[0])
        if Rating1 is None:
            Rating.append(np.nan)
        else:
            Rating.append(Rating1.text.strip()[-3:])
        if Highly_rate1 is None:
            Highly_rated.append(np.nan)
        else:
            Highly_rated.append(Highly_rate1.text.strip())
        if rating_value_span is None:
            Critically_rated.append(np.nan)
        else:
            Critically_rated.append(rating_value_span.text.strip())
    #Page_Number.append(i)
```

```python
"""Since we are unable to extract the 'Founded', 'India count', and 'Global count'
 data directly from the first page, so we are looping through each company's 
 overview page to extract the data. If page isn't found then it will append Null values to lists"""

# required lists are created to store data
Founded=[]
India_count = []
Global_count = []
#li = []

## looping through Company_Name list
for i in Company_Name:
    # i = i.replace('&', 'and')

    # creates an url for each company overview
    url1 = "https://www.ambitionbox.com/overview/"+i.replace(' ','-').replace('&','and')+"-overview"
    response=requests.get(url1,headers={'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Mobile Safari/537.36'})
    page_content = response.text
    print(response.status_code)
    soup = BeautifulSoup(page_content,'html.parser')
    #print(url1)
    #li.append(url1)


    ##finds all elements from the provided class
    found_global_india = soup.find_all('span', class_='css-1jxf684 text-primary-text text-sm font-pn-600 flex-[6] md:flex-[auto]')
    #print(found_global_india)

    ## extracts and appends the data, & handles exception if no data is found
    try:
        Founded.append(found_global_india[0].text)
    except Exception as E:
        Founded.append(np.NaN)
    try:
        India_count.append(found_global_india[1].text)
    except Exception as E:
        India_count.append(np.NaN)
    try:
        Global_count.append(found_global_india[2].text)
    except Exception as E:
        Global_count.append(np.NaN)
#data = pd.DataFrame(li,columns=['url'])"""
```
